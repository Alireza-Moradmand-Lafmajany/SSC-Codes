```python
import pyfluids 
from pyfluids import Fluid, Mixture, FluidsList, Input, HumidAir, InputHumidAir
```

```python
import numpy as np
```

```python
Air = Fluid(FluidsList.Air)
point_1 = Air.with_state(Input.temperature(24),Input.pressure(101000))
point_1.entropy
```

```python
def k2c (temp):
  temp = temp - 273.15
  return temp


s = []
p = []
v =[]

for i in np.linspace(point_1.specific_volume, point_1.specific_volume/8 , 100):
  point = Air.with_state(Input.specific_volume(i), Input.entropy(point_1.entropy))
  p.append(point.pressure/100000)
  v.append(point.specific_volume)
  if i == point_1.specific_volume/8 :
   point_2 = Air.with_state(Input.specific_volume(i), Input.entropy(point.entropy))



```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(v, p, label='P-V plot', color='blue')
```
